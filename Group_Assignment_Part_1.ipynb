## Idea 1: Online Music Streaming Service (Spotify/Apple Music)
Our idea is to analyze the data found in music streaming apps to discover what song(s) are the most popular. There is always a need to find more great music, and our streaming service personalizes songs for users tailored to their own taste. We are approaching the analytics of how to recommended songs differently than others by finding many similar songs with shared common data (such as if the song artist’s ages and genres are the same for two different artists). By looking at different aspects of the users music library, such as the year the song was made and the artist age, our idea is able to recommend more tailored songs. This idea is unique because it is something that many people in college are using and interested in. This is not a brand-new product or business, we are borrowing from the creators of music streaming services such as Spotify and Apple Music. A typical business interaction would be an external model between the app and user. ## Idea 2: GrubHub
Our idea is to determine the most popular restaurants on GrubHub in order to create a selection menu based on the user’s location and preference. In order to create this list, multiple components will be utilized to assure the restaurants they initially view are the ones that are the most appealing to the specific user. What makes it unique is that this database will be able to accurately organize a wide variety of local restaurants into ones that would appeal to the user the most, as it can potentially improve the business strategy of GrubHub, allowing for improvement of engagement with their customers. This is not an original idea, as we are borrowing the idea from GrubHub, which connects diners with local restaurants in order to have their food delivered right to their location. Many of the transactions performed by this company are the same. We are attempting to improve the user interface of the application, not the worker’s side of the transaction. The user opens the app and inserts their address or current location. From there, the app creates a list of local restaurants they have partnered with to deliver food to the user, ranging in a wide variety of cuisines. Once selecting a restaurant, the restaurant’s menu opens up to allow people to select and even customize their order. After that, the order will be placed and a GrubHub worker will be notified of the order and accept it. The worker will proceed to go to the restaurant, retrieve the food, and drive it over to the location of the user for drop-off.

## Idea 3: Netflix
This idea involves analyzing the streaming services of Netflix to find the most popular categories of shows and movies in order to create more accurate content suggestions oriented towards their audience. We can implement the concept through comparing factors of data beyond just those of the movies themselves such as ratings or the time of day the program is typically viewed. What makes this idea unique is it allows Netflix to make recommendation on movies and shows tailor-made for the viewer directly based on their interests. This result would ultimately enhance the user experience along with the overall quality of the services being provided. We would be borrowing the already existing product of Netflix's streaming services in order to realize the goal of determining what forms of content the end user truly desires. A typical business transaction at this company is relatively simple in nature. The user, who needs to have signed up for an account, logs in to the application and selects the program they wish to view from the extensive database of movies and shows. After the selection is viewed, it is stored in either a recently watched or continue watching list that is conveniently placed on the first screen of the application for the user to see. Netflix then uses this data to make recommendations of movies and shows that they believe the user would enjoy based on their watch history.
