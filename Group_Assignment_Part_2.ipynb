
**General:** Our idea is to determine the most popular restaurants on GrubHub in order to create a specific menu for each individual customer. The selection menu is based on the user’s location and dietary preferences. In order to create this list, multiple components will be utilized to ensure the restaurant's initial menu is the one that is the most appealing to the specific user. What makes our idea unique compared to other delivery apps is the level of customization that we will offer. This database will be able to accurately organize a wide variety of local restaurants into ones that would appeal to the user the most, as it can potentially improve the business strategy of GrubHub, allowing for improvement of engagement and introducing a new audience with customers. This is not an original idea, as we are borrowing the idea from GrubHub, which connects diners with local restaurants in order to have their food delivered right to their location. Many of the transactions performed by this company are the same. We are attempting to improve the user interface of the application, not the worker’s side of the transaction. The user opens the app and inserts their address or current location. From there, the app creates a list of local restaurants they have partnered with to deliver food to the user, ranging in a wide variety of cuisines. Once selecting a restaurant, the menu opens up with our customizations to allow people to select and even personalize their order. After that, the order will be placed and a GrubHub worker will be notified of the request and accept it. The worker will then proceed to go to the restaurant, retrieve the food, and drive it over to the location of the user for drop-off. 

**Updated Attribute**: We chose to focus on the attribute of caloric breakdown of each food item offered in conjunction with its restaurant creator. The goal is for this added feature to allow for better user capability and provide as much information as possible to inform the user to the best of our ability. Our emphasis when discussing the caloric breakdown is directed at the time of day that the user is planning to place their order (breakfast, lunch, dinner). This feature will let the user more accurately plan their caloric intake for the day based on a 2000 calorie scale. Showing the calories for each item ordered will be an automatic function opposed to background information. The calories will be able to be extended if clicked on to show the full details of nutritional information (calories from fat, sodium, sugars, etc.). After culminating caloric information, food choice information, and time of day ordered, our function will be able to better recommend new options that the user may be interested in.  

**Updated Constraints**: A factor of the data that we feel is important to implement is a constraint for the amount of time that the process of delivering the food will consume. We will need to limit the available data in order to prevent unreasonable outliers from occuring. The time between ordering and receiving the product will need to stay within a reasonable range to avoid the possibility of the user selecting food from a restaurant much further away then they intended. Our reasonable range of time is 1 hour. This waiting period is typically a result of the distance from the eatery to the customer and an additional delivery fee is determined based on this information. The main purpose of the constaint is to protect the user from incurring an unnecessarily long delay as well as a hefty service charge that is less than desirable to pay.

**Part One Details:**

The details stored in every table are the attributes that we will use in order to create the dynamic data for the end-user. There are multiple components that will be taken into consideration to output recommended restaurants for GrubHub users: time of order, price, estimated time of arrival, cuisine type, and calories. Our “end-user” data is the top restaurants that the app believes the specific user will enjoy the most in their respective location. 

<b>Entities (Attributes):</b> Restaurant (Restaurant Name, Time of Order), Food Items (Item Name, Food Price), Delivery Details (Delivery Fee, Time Until Arrival), Food Details (Food Category, Calories)

<b>Constraint:</b> Outlier for Time Until Arrival as distance can play a huge part in delivery fee.

<b>Goal:</b>The goal of our project is to assure that the application picks restaurants for the user based on personal preference and by taking several other crucial components (see above) into account.
